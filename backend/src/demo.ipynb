```python
from tldw import TLDW

tldw_model = TLDW(pretrained_whisper_model="base")  # new TLDW instance
```

```python
def present(summary):
    print(summary["output_text"])

    out = summary["output"]
    print(f"OUTPUT: {out}")
    for highlight in out.gpt3_highlights:
        print(highlight.__str__())

    visualization = summary["visualization"]

    if visualization is not None:
        print(visualization)

        img_url = summary["visualization"]["data"][0]["url"]
        response = requests.get(img_url)
        img = Image.open(BytesIO(response.content))

        return img


    return None
```

```python
# Example 1: Meet the Pine-Sol® Family
summary_1 = tldw_model.summarize_yt_video("https://www.youtube.com/watch?v=J1uIU5t1fMo", visualize=True)

print(summary_1["output"])
present(summary_1)
```

```python
# Example 2: The danger of silence | Clint Smith
summary_2 = tldw_model.summarize_yt_video("https://www.youtube.com/watch?v=NiKtZgImdlY", num_highlights=10, visualize=False)

print(summary_2["output"])
present(summary_2)
```

```python
# Example 3: Does Bosnia have the most complicated political system in Europe?
summary_3 = tldw_model.summarize_yt_video("https://youtu.be/sCEYDdMue-8", visualize=True)

print(summary_3["output"])
present(summary_3)
```
